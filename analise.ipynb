```python
import pandas as pd

df_decadas = pd.read_csv("terror_decadas.csv")
df_subgeneros = pd.read_csv("top10_subgeneros.csv")
```

```python
df_decadas_limpo = df_decadas[[
    "title",
    "release_date",
    "vote_average",
    "popularity",
    "decada"
]].copy()

# Criar coluna apenas com o ano
df_decadas_limpo["ano"] = df_decadas_limpo["release_date"].str[:4]

df_decadas_limpo.head()
```

```python
df_decadas_limpo = df_decadas[[
    "title",
    "release_date",
    "vote_average",
    "popularity",
    "decada"
]].copy()

df_decadas_limpo["ano"] = df_decadas_limpo["release_date"].str[:4]

df_decadas_limpo.head()
```

```python
df_decadas_limpo.groupby("decada")[["vote_average", "popularity"]].mean()
```

```python
import matplotlib.pyplot as plt

contagem = df_decadas_limpo["decada"].value_counts().sort_index()

plt.figure(figsize=(8,4))
contagem.plot(kind="bar")
plt.title("Quantidade de filmes de terror por década")
plt.xlabel("Década")
plt.ylabel("Quantidade de filmes")
plt.show()
```

```python
medias = df_decadas_limpo.groupby("decada")["vote_average"].mean()

plt.figure(figsize=(8,4))
medias.plot(kind="line", marker="o")
plt.title("Nota média dos filmes de terror por década")
plt.xlabel("Década")
plt.ylabel("Nota média")
plt.grid(True)
plt.show()
```

```python
pop = df_decadas_limpo.groupby("decada")["popularity"].mean()

plt.figure(figsize=(8,4))
pop.plot(kind="line", marker="o", color="green")
plt.title("Popularidade média dos filmes de terror por década")
plt.xlabel("Década")
plt.ylabel("Popularidade média")
plt.grid(True)
plt.show()
```
